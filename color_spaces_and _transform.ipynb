```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
imageRGB = cv2.cvtColor(cv2.imread('images/water_balloons.jpg'), cv2.COLOR_BGR2RGB)
imageHSV = cv2.cvtColor(cv2.imread('images/water_balloons.jpg'), cv2.COLOR_BGR2HSV)
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize = (20, 10)) 
ax1.set_title('RED')
ax2.set_title('GREEN')
ax3.set_title('BLUE')

ax1.imshow(imageRGB[:,:,0], cmap = 'gray')
ax2.imshow(imageRGB[:,:,1], cmap = 'gray')
ax3.imshow(imageRGB[:,:,2], cmap = 'gray')
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize = (20, 10)) 
ax1.set_title('HUE')
ax2.set_title('SATURATION')
ax3.set_title('VALUE')

ax1.imshow(imageHSV[:,:,0], cmap = 'gray')
ax2.imshow(imageHSV[:,:,1], cmap = 'gray')
ax3.imshow(imageHSV[:,:,2], cmap = 'gray')
```

```python
upper_pink = np.array([255, 255, 230])
lower_pink = np.array([180, 0, 100])
mask = cv2.inRange(imageRGB, lower_pink, upper_pink)

masked_image = np.copy(imageRGB)
masked_image[mask == 0] = [0,0,0]
plt.imshow(masked_image, cmap = 'gray')
```

```python
upper_hue = np.array([180, 255, 255])
lower_hue = np.array([160, 0, 0])
mask = cv2.inRange(imageHSV, lower_hue, upper_hue)

masked_image = np.copy(imageRGB)
masked_image[mask == 0] = [0,0,0]
plt.imshow(masked_image, cmap = 'gray')
```
