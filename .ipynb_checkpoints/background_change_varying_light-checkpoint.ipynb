```python
%matplotlib inline
image = cv2.cvtColor(cv2.imread('images/pizza_bluescreen2.jpg'), cv2.COLOR_BGR2HSV)
plt.imshow(image)
```

```python
%matplotlib inline
f, (ax1,ax2,ax3) = plt.subplots(1,3, figsize = (20, 6))
ax1.imshow(image[:,:,0])
ax2.imshow(image[:,:,1])
ax3.imshow(image[:,:,2])
```

```python
upper = np.array([114, 255, 255])
lower = np.array([90, 0, 0])
mask = cv2.inRange(image, lower, upper)
plt.imshow(mask, cmap = 'gray')
```

```python
image[mask != 0] = [0,0,0]
masked_image = cv2.cvtColor(image, cv2.COLOR_HSV2RGB)
plt.imshow(masked_image)
```

```python
back_img = cv2.cvtColor(cv2.imread('images/space_background.jpg'), cv2.COLOR_BGR2RGB)
back_img = back_img[0:image.shape[0], 0:image.shape[1]]
back_img[mask == 0 ] = [0,0,0]
plt.title('Background Image')
plt.imshow(back_img)
```

```python
final_img = back_img + masked_image
plt.imshow(final_img)
```
